<h1 style="text-align: center;">Polynomial Regression</h1>

Polynomial Regression is a regression algorithm that models the relationship between a dependent(y) and independent variable(x) as nth degree polynomial. The Polynomial Regression equation is given below:
<p style="text-align: center;">y= b0+b1x1+ b2x12+ b2x13+...... bnx1</p>

Polynomial regression is a type of regression analysis used in statistics and machine learning when the relationship between the independent variable (input) and the dependent variable (output) is not linear. While simple linear regression models the relationship as a straight line, polynomial regression allows for more flexibility by fitting a polynomial equation to the data.

When the relationship between the variables is better represented by a curve rather than a straight line, polynomial regression can capture the non-linear patterns in the data.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 35)
y = [4, 5, 6, 8, 10, 9, 11, 15, 22, 28, 33, 37, 40,
     45, 50, 55, 57, 63, 69, 72, 75, 80, 88, 100, 115,
     125, 141, 150, 165, 175, 186, 201, 210, 226, 243]

plt.scatter(x, y, color = "black")
plt.show()
```

```python
lin = LinearRegression()
lin.fit(x.reshape(-1, 1), y)

def linreg():
    plt.scatter(x, y, color = "black")
    plt.plot(x, lin.predict(x.reshape(-1, 1)), color = "magenta")
    plt.title("Linear regression")
    plt.xlabel("x values")
    plt.ylabel("y values")
    plt.show()
    return
linreg()
```

```python
print("Only simple linear regression - value of R_2 is: ", r2_score(y, lin.predict(x.reshape(-1, 1))))
```

```python
#degree of the polynomial (degree = 2)
poly = PolynomialFeatures(degree = 3, include_bias = False)
poly_features = poly.fit_transform(x.reshape(-1, 1))
```

```python
poly_features[0:5]
```

```python
poly_reg_model = LinearRegression()
```

```python
poly_reg_model.fit(poly_features, y)
```

```python
y_pred = poly_reg_model.predict(poly_features)
y_pred
```

```python
plt.scatter(x, y, color = "black")
plt.plot(x, y_pred, c = "red")
plt.show()
```

```python
print("With polynomial regression - value of R_2 is: ", r2_score(y, y_pred))
```
